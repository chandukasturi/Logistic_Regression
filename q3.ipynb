```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import classification_report,recall_score,precision_score

batch, tb = tf.keras.datasets.mnist.load_data()


X_train = batch[0]  
y_train = batch[1]


indices = np.argwhere(y_train < 5)

y_train=y_train[indices]
y_train = np.reshape(y_train,(30596))
X_train=X_train[indices]
X_train=X_train.reshape(30596,28,28)


X_test = tb[0]  
y_test = tb[1]

indices1 = np.argwhere(y_test < 5)
y_test=y_test[indices1]
y_test = np.reshape(y_test,(5139))
X_test=X_test[indices1]
X_test=X_test.reshape(5139,28,28)

y_predict = X_test


X_train_new = X_train / 255.0
X_test_new = X_test / 255.0
X_train_new[0]

x_valid,X_train = X_train_new[:5000],X_train_new[5000:]
y_valid,Y_train = y_train[:5000],y_train[5000:]
X_test = X_test_new


print("ANN with 1 Hidden layer using sigmoid activation function")

np.random.seed(40)
tf.random.set_seed(40)

model_sigmoid = keras.models.Sequential()
model_sigmoid.add(keras.layers.Flatten(input_shape=[28,28]))
model_sigmoid.add(keras.layers.Dense(512, activation="sigmoid"))
model_sigmoid.add(keras.layers.Dense(10,activation="softmax"))
model_sigmoid.summary()
model_sigmoid.compile(loss="sparse_categorical_crossentropy",optimizer="sgd",metrics=["accuracy"])
model_sigmoid_fit = model_sigmoid.fit(X_train,Y_train,epochs=5,validation_data=(x_valid,y_valid))
model_sigmoid_fit.params
model_sigmoid_fit.history
weights, biases = model_sigmoid.layers[1].get_weights()
model_sigmoid.evaluate(X_test,y_test)

y_pred = model_sigmoid.predict_classes(X_test)
print(classification_report(y_test, y_pred))





print("ANN with 1 Hidden layer using relu activation function")

np.random.seed(40)
tf.random.set_seed(40)
model_relu = keras.models.Sequential()
model_relu.add(keras.layers.Flatten(input_shape=[28,28]))
model_relu.add(keras.layers.Dense(512, activation="relu"))
model_relu.add(keras.layers.Dense(10,activation="softmax"))
model_relu.summary()
weights, biases = model_relu.layers[1].get_weights()
model_relu.compile(loss="sparse_categorical_crossentropy",optimizer="sgd",metrics=["accuracy"])
model_relu_fit = model_relu.fit(X_train,Y_train,epochs=5,validation_data=(x_valid,y_valid))
model_relu_fit.params
model_relu_fit.history
model_relu.evaluate(X_test,y_test)
y_pred = model_relu.predict_classes(X_test)
print(classification_report(y_test, y_pred))


```
